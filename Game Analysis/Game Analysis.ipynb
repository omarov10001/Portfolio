## Part 1

THe first task goal is to create an ETL script using Python and SQL that extracts the raw data from “sample_data” and creates a flat events table.

The first table will be named “events_flat_omar”. The “events_flat” table will be organized so that each row in the raw data table is transformed into a row in the “events_flat” table and so that all data in “eventProperties” column in the raw data table is extracted and put as a separate column in the “events_flat” table for easier processing and querying.

```python
credentials_path = "C:/Users/HP/Desktop/bigquery_creds.json"
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = credentials_path

bq_project = "ng-bq-test"
bq_dataset = "ng_test"
client = bigquery.Client(project=bq_project)
dataset_ref = client.dataset(bq_dataset)


sql_query_table1 = """

CREATE TABLE `ng-bq-test.ng_test.events_flat_omar` AS 

SELECT
 userID
,eventUUID
,clientVersion
,platform
,eventDate
,eventTimestamp
,eventName
,eventProperties
,JSON_EXTRACT(eventProperties, '$.supporID') AS supportID
,JSON_EXTRACT(eventProperties, '$.advertisingID') AS advertisingID
,JSON_EXTRACT(eventProperties, '$.sessionID') AS sessionID
,JSON_EXTRACT(eventProperties, '$.sessionCount') AS sessionCount
,JSON_EXTRACT(eventProperties, '$.userRank') AS userRank
,JSON_EXTRACT(eventProperties, '$.userXP') AS userXP
,JSON_EXTRACT(eventProperties, '$.userLevel') AS userLevel
,JSON_EXTRACT(eventProperties, '$.teamName') AS teamName
,JSON_EXTRACT(eventProperties, '$.teamID') AS teamID
,JSON_EXTRACT(eventProperties, '$.teamRank') AS teamRank
,JSON_EXTRACT(eventProperties, '$.teamLiveRank') AS teamLiveRank
,JSON_EXTRACT(eventProperties, '$.sdkVersion') AS sdkVersion
,JSON_EXTRACT(eventProperties, '$.associationName') AS associationName
,JSON_EXTRACT(eventProperties, '$.associationID') AS associationID
,JSON_EXTRACT(eventProperties, '$.softCurrencyBalance') AS softCurrencyBalance
,JSON_EXTRACT(eventProperties, '$.hardCurrencyBalance') AS hardCurrencyBalance
,JSON_EXTRACT(eventProperties, '$.eventDate') AS eventDate_Json
,JSON_EXTRACT(eventProperties, '$.userStartDate') AS userStartDate
,JSON_EXTRACT(eventProperties, '$.uaChannel') AS uaChannel
,JSON_EXTRACT(eventProperties, '$.deviceName') AS deviceName
,JSON_EXTRACT(eventProperties, '$.deviceType') AS deviceType
,JSON_EXTRACT(eventProperties, '$.os') AS os
,JSON_EXTRACT(eventProperties, '$.osVersion') AS osVersion
,JSON_EXTRACT(eventProperties, '$.timezoneOffset') AS timezoneOffset
,JSON_EXTRACT(eventProperties, '$.serverEnv') AS serverEnv
,JSON_EXTRACT(eventProperties, '$.clientVersion') AS clientVersion_Json
,JSON_EXTRACT(eventProperties, '$.platform') AS platform_Json
,JSON_EXTRACT(eventProperties, '$.playerCountry') AS playerCountry
,JSON_EXTRACT(eventProperties, '$.deviceManufacturer') AS deviceManufacturer
,JSON_EXTRACT(eventProperties, '$.vendorId') AS vendorId
,JSON_EXTRACT(eventProperties, '$.localeTag') AS localeTag
,JSON_EXTRACT(eventProperties, '$.botAmount') AS botAmount
,JSON_EXTRACT(eventProperties, '$.mapName') AS mapName
,JSON_EXTRACT(eventProperties, '$.playerAmount') AS playerAmount
,JSON_EXTRACT(eventProperties, '$.gameMode') AS gameMode
,JSON_EXTRACT(eventProperties, '$.matchmakingTime') AS matchmakingTime
,JSON_EXTRACT(eventProperties, '$.isTutorial') AS isTutorial
,JSON_EXTRACT(eventProperties, '$.loadoutTrophies') AS loadoutTrophies
,JSON_EXTRACT(eventProperties, '$.blasterOne') AS blasterOne
,JSON_EXTRACT(eventProperties, '$.blasterTwo') AS blasterTwo
,JSON_EXTRACT(eventProperties, '$.blasterThree') AS blasterThree
,JSON_EXTRACT(eventProperties, '$.powerCard') AS powerCard
,JSON_EXTRACT(eventProperties, '$.playerId') AS playerId
,JSON_EXTRACT(eventProperties, '$.sensitivityGeneral') AS sensitivityGeneral
,JSON_EXTRACT(eventProperties, '$.sensitivity_X') AS sensitivity_X
,JSON_EXTRACT(eventProperties, '$.sensitivity_Y') AS sensitivity_Y
,JSON_EXTRACT(eventProperties, '$.lookingSensitivityX') AS lookingSensitivityX
,JSON_EXTRACT(eventProperties, '$.lookingSensitivityY') AS lookingSensitivityY
,JSON_EXTRACT(eventProperties, '$.friendCount') AS friendCount
,JSON_EXTRACT(eventProperties, '$.isInLobby') AS isInLobby

FROM `ng-bq-test.ng_test.sample_data`
"""

exec_query = client.query(sql_query_table1)
result = exec_query.result()
table = result.to_dataframe()
table
```

## Part 2 

The second part goal is to create a separate table for the “matchStart” matchStart. This will be done in a similar fashion as the creation of events_flat table. Instead of using a raw data table, I will use the events_flat table I created in the first part and take all necessary event properties (=columns) with the events to the new table.


```python
sql_query_table2 = """
CREATE TABLE `ng-bq-test.ng_test.matchStart_omar` as

SELECT 
userID,eventUUID,gameMode,blasterOne,blasterTwo,blasterThree
FROM `ng-bq-test.ng_test.events_flat_omar` 
WHERE eventName = 'matchStart'
"""

exec_query_table2 = client.query(sql_query_table2)
result_table2= exec_query_table2.result()
table2 = result_table2.to_dataframe()
table2
```
